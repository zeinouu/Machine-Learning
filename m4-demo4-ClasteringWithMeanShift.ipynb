```python
import pandas as pd
```

```python
titanic_data = pd.read_csv('Documents/titanic.csv', quotechar='"')
titanic_data.head()
```

```python
columns_to_drop = ['Passengerid', 'zero', 'zero.1', 'zero.1', 'zero.2', 'zero.3', 'zero.4', 'zero.5', 'zero.6', 'zero.7', 'zero.8', 'zero.9', 'zero.10', 'zero.11', 'zero.12', 'zero.13', 'zero.14', 'zero.15', 'zero.16', 'zero.17', 'zero.18']
titanic_data.drop(columns=columns_to_drop, inplace=True)
titanic_data.head()
```

```python
titanic_data = pd.get_dummies(titanic_data, columns=['Embarked'])
titanic_data.head()
```

```python
titanic_data.replace({True: 1, False: 0}, inplace=True)
```

```python
from sklearn.cluster import MeanShift

analyzer = MeanShift(bandwidth=50)
analyzer.fit(titanic_data)
```

there is a function that estimates the bandwidth :

```python
from sklearn.cluster import estimate_bandwidth
estimate_bandwidth(titanic_data)
```

```python
import numpy as np
```

Now we see how many cluster we have :

```python
labels = analyzer.labels_
```

```python
import numpy as np
np.unique(labels)
```

Now we will add a new column called cluster group to our data frame. This column will contain information as to what cluster a particular passenger belongs to.

```python
import numpy as np

titanic_data['cluster_group'] = np.nan
data_length = len(titanic_data)
for i in range(data_length):
    titanic_data.iloc[i, titanic_data.columns.get_loc('cluster_group')] = labels[i]
```

```python
titanic_data.head()
```

```python
titanic_data.describe()
```

Now we will regroupe the clusters by using the cluster group column that we just added and then find the mean for each cluster.
This will be the average of all characteristics within each group.

```python
titanic_cluster_data = titanic_data.groupby(['cluster_group']).mean()
titanic_cluster_data
```

The average informations will be much valuable if we add the number of samples that exist in each cluster.
Now we will add a counts column to our data frame and set its value to the size of each cluster.

```python
titanic_cluster_data['Counts'] = pd.Series(titanic_data.groupby(['cluster_group']).size())
titanic_cluster_data
```
